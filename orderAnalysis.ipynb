```python
import pandas as pd
orders=pd.read_csv("orders.csv",na_values=['Not Available', 'unknown'])
orders.head()
```

```python
orders.info()
```

```python
orders["Ship Mode"].unique()
```

```python
orders.columns=orders.columns.str.lower()
orders.columns=orders.columns.str.replace(" ","_")
orders.columns
```

```python
orders['discount_amount']=orders['list_price']*orders['discount_percent']/100
orders.head()
```

```python
orders['sales_price']=orders['list_price']-orders['discount_amount']
orders.head()
```

```python
orders['profit']=orders['sales_price']-orders['cost_price']
orders.head()
```

```python
orders['order_date']=pd.to_datetime(orders['order_date'],format='%Y-%m-%d')
orders.dtypes
```
